import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

wygenerowanie danych

```python
data={
    'size':['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight':[500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

Tworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

zmiana typu danych i wstępna eksploracja

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')
df.info()
```

STANDARYZACJA Standard Scaler std()- pandas nieobciążony, std() - numpy - obciążony

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean())/ df['price'].std()
```

Z funkcją

```python
def standardize(x):
    return (x - x.mean()) / x.std()

standardize(df['price'])
```

```python
from sklearn.preprocessing import scale
scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler
scale(df['price'])
scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```
