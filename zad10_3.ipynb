Wersja z trzema wykresam na raz, zbyt duży rozrzut odchylenia żeby dokładnie to zobaczyć

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def gauss(x, sigma=1, mu=3):
    return stats.norm.pdf(x, mu, sigma)

x = np.linspace(-5, 9, 10000)
plt.figure(figsize=(12, 8))
plt.plot(x, gauss(x, sigma=1), label='odchylenie standardowe = 1')
plt.plot(x, gauss(x, sigma=100), label='odchylenie standardowe = 100')
plt.plot(x, gauss(x, sigma=10000), label='odchylenie standardowe = 10000')

plt.legend(loc='upper left')
plt.show()
```
