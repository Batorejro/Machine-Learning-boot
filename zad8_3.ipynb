```python
lata = df['Rok'].unique()
result = []
for year in lata:
    if year >= 1960 and year <= 1990:
        result.append(year)
print(result)
print("Pomiędzy 60 a 90 rokiem, zastało wydanych:", len(result), "albumów")
```
