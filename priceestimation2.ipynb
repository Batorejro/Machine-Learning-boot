```python
import numpy as np
import matplotlib.pyplot as plt

# Dane wejściowe
initial_value = 90000  # Początkowa kwota
target_value = 120000
years = 5
rates = [0.06, 0.12]

# Wartość mieszkania za 5 lat
future_values = [round(initial_value * (1 + rate) ** years, 2) for rate in rates]

print("Wartość po 5 latach:", future_values)

# Wykres wzrostu mieszkania w zależności od stopy wzrostu
years_range = np.arange(1, 21)  # Zakres lat od 1 do 20

for rate in rates:
    values = [initial_value * (1 + rate) ** year for year in years_range]
    plt.plot(years_range, values, label=f'Stopa wzrostu: {rate*100}%')

plt.axhline(y=target_value, color='g', linestyle=':', label='Cel: 120000')
plt.xlabel('Lata')
plt.ylabel('Wartość mieszkania')
plt.title('Wzrost wartości mieszkania w czasie')
plt.legend()
plt.grid(True)
plt.show()
```
