```python
petal = iris['petal_length'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(petal.index, petal, color='green')
plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
sns.set_style("darkgrid")

# Identify numerical columns
numerical_columns = iris.select_dtypes(include=["int64", "float64"]).columns


plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(iris[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(iris[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.swarmplot(x="sepal_length", y="petal_length", data=iris, palette='viridis')

plt.title('kwiatki w kropki')
plt.xlabel('S_L')
plt.ylabel('P_L')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_iris = iris.select_dtypes(include=['float', 'int'])
plt.figure(figsize=(15, 10))


sns.heatmap(numerical_iris.corr(), annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)

plt.title('Correlation Heatmap')
plt.show()
```
